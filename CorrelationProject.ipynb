```python
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as stats
import math
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\shuru\OneDrive\Desktop\moviesproject\movies.csv')

```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df.dtypes
```

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
df['gross'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
```

```python
df['released'] = df['released'].astype(str)

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})')

df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending = False)
```

```python
pd.reset_option('display.max_columns', 0)
```

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings Per Film')
plt.xlabel('Budget for Film(HM)')
plt.ylabel('Gross Earning(B)')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > .5]
high_corr
```
